```python
import matplotlib.pyplot as plt #this is subpackage for ploting( pyplot)
import numpy as np
```

```python
arr=np.arange(10)
```

```python
arr
```

```python
arrsq=arr**2
```

# Line Graph

```python
plt.plot([1,4,9],[2,5,3])
plt.show()
```

```python
#esc+l show u line number
plt.plot(arr,arr,marker="^",color="red",label="y=x") #try start *,o
#plt.show() #will make two different graph
#basically we got graph at pt where we do plt.show()
plt.plot(arr,arrsq,marker="*",color="black",label="y=x**2") #arr should be on x axis and arrsq on y axis
plt.legend()
plt.xlabel("Numbers(0-10)")
plt.ylabel("range of function")
plt.title("Comparison function")
plt.xlim(1,20)
#plt.xticks()
#plt.
#plt.ylim
plt.show()
```
